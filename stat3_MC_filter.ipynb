```python
from rdkit import Chem
import pandas as pd

def calculate_mce18(smiles):
    mol = Chem.MolFromSmiles(smiles)
    if mol is None:
        return None
    num_atoms = mol.GetNumAtoms()
    num_bonds = mol.GetNumBonds()
    num_rings = mol.GetRingInfo().NumRings()
    num_H = sum(atom.GetTotalNumHs() for atom in mol.GetAtoms())
    mce18 = num_atoms + num_bonds + 2 * num_rings - num_H
    return mce18

```

```python
import sys
import os
# sys.path.append(os.path.join(os.environ['CONDA_PREFIX'],'share','RDKit','Contrib'))

from rdkit.Contrib.NP_Score import npscorer
fscore = npscorer.readNPModel()
from rdkit import Chem
def calculate_np_score(smiles):
    mol = Chem.MolFromSmiles(smiles)
    if mol is None:
        return None  # Hoặc giá trị mặc định nếu SMILES không hợp lệ
    try:
        np_score = npscorer.scoreMol(mol, fscore)
        return np_score
    except Exception as e:
        print(f"Lỗi khi tính NP_Score cho {smiles}: {e}")
        return None
```

```python
df["Smiles"] = df["Smiles"].fillna("").astype(str)
```

```python
from tqdm.auto import tqdm
from rdkit.Chem.rdMolDescriptors import CalcNumHBA,CalcNumHBD,CalcFractionCSP3,CalcTPSA
from rdkit.Chem.Descriptors import MolWt
from rdkit.Chem.Crippen import MolLogP
import sascorer
from sascorer import calculateScore

tqdm.pandas()
df['ROMol'] = df.Smiles.progress_apply(Chem.MolFromSmiles)
df['MW'] = df.ROMol.progress_apply(MolWt)
df['LogP'] = df.ROMol.progress_apply(MolLogP)
df['HBA'] = df.ROMol.progress_apply(CalcNumHBA)
df['HBD'] = df.ROMol.progress_apply(CalcNumHBD)
df['PSA'] = df.ROMol.progress_apply(CalcTPSA)
df['Fsp3'] = df.ROMol.progress_apply(CalcFractionCSP3)
df['SA'] = df.ROMol.progress_apply(calculateScore)
df['MCE18'] = df['Smiles'].apply(calculate_mce18)
df['NP_Score'] = df['Smiles'].apply(calculate_np_score)


df.head(10)
```

```python


import pandas as pd

# --- 1. Hàm tính số vi phạm Lipinski ---
def lipinski_violations(row):
    violations = 0
    if row["HBA"] > 10:
        violations += 1
    if row["HBD"] > 5:
        violations += 1
    if row["MW"] > 600:
        violations += 1
    if row["LogP"] > 5:
        violations += 1
    return violations

df["Lipinski_Pass"] = df.apply(lipinski_violations, axis=1)


# --- 2. Điều kiện Pfizer 3/75 ---
def pfizer_pass(row):
    return not (row["LogP"] >= 3 and row["PSA"] <= 75)

df["Pfizer_pass"] = df.apply(pfizer_pass, axis=1)


# --- 3. Các điều kiện định lượng ---
df["Fsp3_pass"] = df["Fsp3"] >= 0.42
df["MCE_pass"]  = df["MCE18"] >= 45
df["SA_pass"]   = df["SA"] <= 6
df["NP_pass"]   = df["NP_Score"].between(-5, 5, inclusive="both")


# --- 4. Tổng hợp điều kiện cuối (không bộ lọc PAINS/BMS/Chelator) ---
df["Final_pass"] = (
    (df["Lipinski_Pass"] <= 1) &
    (df["Pfizer_pass"]) &
    (df["Fsp3_pass"]) &
    (df["MCE_pass"]) &
    (df["SA_pass"]) &
    (df["NP_pass"])
)
df['Lipinski_Pass'] = df['Lipinski_Pass'] <= 1

# --- 5. Lọc phân tử đạt điều kiện ---
df_filtered = df[df["Final_pass"]].copy()


df["NP_pass"]   = df["NP_Score"].between(-5, 5, inclusive="both")

df_filtered['Lipinski_Pass'] = df_filtered['Lipinski_Pass'] <= 1
df_filtered.drop(columns=['ROMol'],inplace =True)
df_filtered
#df_filtered.to_csv('/content/drive/MyDrive/stat3/hehe1.csv',index=False)
```
